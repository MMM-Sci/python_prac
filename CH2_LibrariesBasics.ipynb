```python
# p.48-49

matrix = np.array([[1,-1,-1], [-1,1,-1], [-1,-1,1]])

print("행렬식   linalg.det(matrix)")
print(linalg.det(matrix))
# cf. https://m.blog.naver.com/PostView.naver?isHttpsRedirect=true&blogId=mykepzzang&logNo=221080004072

print("\n역행렬   linalg.inv(matrix)")
print(linalg.inv(matrix))

print("\n원래 행렬과 역행렬의 곱 = 단위 행렬")
print(matrix.dot(linalg.inv(matrix)))

print("\n고윳값과 고유벡터")
eig_value, eig_vector = linalg.eig(matrix)
print(eig_value)
print(eig_vector)
# cf. https://darkpgmr.tistory.com/105#
```

```python
# 2장 종합문제

import math

df = pd.DataFrame([])
df['x']    = np.random.uniform(0, 1, 10000)
df['y']    = np.random.uniform(0, 1, 10000)

for i in range(df.shape[0]):
    df.loc[i, 'norm'] =  math.hypot(df.loc[i, 'x'], df.loc[i, 'y'])

inside = sum(df['norm'] < 1)
outside = sum(df['norm'] >= 1)
print("원 안에 있는 점 개수", inside)
print("원 밖에 있는 점 개수", outside)

print("원주율", 4*inside/10000)

plt.figure(figsize = (10, 10))
plt.scatter(df[df['norm'] < 1]['x'], df[df['norm'] < 1]['y'], color = 'g')
plt.scatter(df[df['norm'] >= 1]['x'], df[df['norm'] >= 1]['y'], color = 'r')
plt.grid(True)
```
